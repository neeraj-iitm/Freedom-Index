```python
import numpy as np
import pandas as pd
from tabula import read_pdf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = read_pdf('hfi2020.pdf',pages='121')
```

```python
df = df[0]
```

```python
df.columns
```

```python
df.rename( columns={'Unnamed: 0':'Indicator'}, inplace=True )
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df
```

```python
def strip(text):
    try:
        return text.replace(" ", "")
    
    except AttributeError:
        return text
```

```python
for ind, column in enumerate(df.columns):
    if ind>0:
        df[column]=df[column].apply(lambda x : strip(x))
```

```python
df = df.reset_index()
```

```python
df = df.drop(index=35,axis=0)
```

```python
df = df.drop('index',axis=1)
```

```python
for ind,column in enumerate(df.columns):
    if ind>0:
        df[column]=pd.to_numeric(df[column])
```

```python
mainIndicatorList = ['Human Freedom',
                     'Economic Freedom',
                     'Rule of Law', 
                     'Security and Safety', 
                     'Movement',
                     'Religious Freedom',
                     'Assoc., Assembly, and Civil Soc.',
                     'Expression and Information',
                     'Identity and Relationships','','']
```

```python
plt.figure(figsize=(15,9))

subset = df.iloc[4:,1:]
y  = sns.heatmap(subset,cmap ="RdYlBu",linewidths=0.005)
plt.yticks(range(df.shape[0]-4),df['Indicator'][4:],rotation=0)

for ind,column in enumerate(df.iloc[4:,0]):
    if column in mainIndicatorList:
        y.get_yticklabels()[ind].set_weight("bold")
plt.show()
```

```python
subset2  = df.iloc[:,1:]
cmap = sns.diverging_palette(0, 230, 90, 45, as_cmap=True)
plt.figure(figsize=(15,9))
y    = sns.heatmap(subset2,cmap =cmap,linewidths=0.005,vmax=10)
plt.yticks(range(df.shape[0]),df['Indicator'],rotation=0)

for ind,column in enumerate(df.iloc[:,0]):
    if column in mainIndicatorList:
        y.get_yticklabels()[ind].set_weight("bold")
plt.show()
```
